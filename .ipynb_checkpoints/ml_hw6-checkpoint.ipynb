```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from torchvision import datasets, transforms
from sklearn.metrics import classification_report
```

```python
dataset_path = 'Housing.csv'
housing_data = pd.read_csv(dataset_path, encoding='utf-8')  # Added explicit encoding
housing_data.head()
```

```python
furnishing_counts = housing_data['furnishingstatus'].value_counts()
unique_counts = furnishing_counts.unique
housing_data.head()
```

```python
housing_encoded = pd.get_dummies(housing_data, columns=['furnishingstatus'], drop_first=False)
housing_encoded = housing_encoded.drop('furnishingstatus_semi-furnished', axis=1)
housing_encoded.head()
```

```python
categorical_features = housing_encoded.select_dtypes(include=['object']).columns
categorical_features
```

```python
housing_encoded[categorical_features] = housing_encoded[categorical_features].apply(lambda x: x.map({'yes': 1, 'no': 0}))
housing_encoded.head()
```

```python
housing_encoded.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
feature_scaler = MinMaxScaler()
target_scaler = MinMaxScaler()
```

```python
features = housing_encoded.drop('price', axis=1)
target = housing_encoded[['price']]
```

```python
features_scaled = pd.DataFrame(feature_scaler.fit_transform(features), columns=features.columns, index=features.index).values
target_scaled = pd.DataFrame(target_scaler.fit_transform(target), columns=target.columns, index=target.index).values
```

```python
features_train, features_test, target_train, target_test = train_test_split(features_scaled, target_scaled, test_size=0.2, random_state=42)
train_features_tensor = torch.tensor(features_train)
train_target_tensor = torch.tensor(target_train)
test_features_tensor = torch.tensor(features_test)
test_target_tensor = torch.tensor(target_test)
```

```python
train_dataset = TensorDataset(train_features_tensor, train_target_tensor)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
```

```python

test_dataset = TensorDataset(test_features_tensor, test_target_tensor)
test_loader = DataLoader(test_dataset, batch_size=32)
```

```python
def train_neural_network(model, loss_function, optimizer, num_epochs, train_loader, test_features, test_targets):
    train_losses = []
    validation_losses = []
    
    for epoch in range(num_epochs + 1):
        # Training phase
        model.train()
        for batch_features, batch_targets in train_loader:
            optimizer.zero_grad()
            predictions = model(batch_features)
            batch_loss = loss_function(predictions, batch_targets)
            batch_loss.backward()
            optimizer.step()
            
        # Validation phase
        model.eval()
        with torch.no_grad():
            test_predictions = model(test_features)
            validation_loss = loss_function(test_predictions, test_targets)
            
            if epoch % 500 == 0:
                print(f'Epoch {epoch}/{num_epochs}, Training Loss: {batch_loss.item()}, Testing Loss: {validation_loss.item()}')
                validation_losses.append(validation_loss)
                train_losses.append(batch_loss)
                
    return train_losses, validation_losses
```

```python
baseline_network = nn.Sequential(
    nn.Linear(train_features_tensor.shape[1], 32),
    nn.Tanh(),
    nn.Linear(32, 1)
).to(torch.float64)

baseline_optimizer = optim.SGD(baseline_network.parameters(), lr=0.02)
baseline_train_loss, baseline_val_loss = train_neural_network(
    baseline_network, 
    nn.MSELoss(), 
    baseline_optimizer, 
    5000, 
    train_loader, 
    test_features_tensor, 
    test_target_tensor
)
```

```python
for layer_name, layer_params in baseline_network.named_parameters():
    print(f"Layer name: {layer_name}, Dimensions: {layer_params.shape}")
```

```python
trainable_params = sum(p.numel() for p in baseline_network.parameters() if p.requires_grad)
print("Total trainable parameters:", trainable_params, '\n')
```

```python
# Deep Neural Network Model
deep_network = nn.Sequential(
    nn.Linear(train_features_tensor.shape[1], 32),
    nn.Tanh(),
    nn.Linear(32, 64),    # Expanded middle layer
    nn.Tanh(),
    nn.Linear(64, 16),    # Bottleneck layer
    nn.Tanh(),
    nn.Linear(16, 1)      # Output projection
).to(torch.float64)

loss_function = nn.MSELoss()
deep_optimizer = optim.SGD(deep_network.parameters(), lr=0.02)
deep_train_loss, deep_val_loss = train_neural_network(
    deep_network, 
    loss_function, 
    deep_optimizer, 
    5000, 
    train_loader, 
    test_features_tensor, 
    test_target_tensor
)
```

```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset, random_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import time
```

```python
import pandas as pd

# Define data source location
cancer_dataset_path = 'cancer (2).csv'

# Import cancer data into pandas dataframe
cancer_data = pd.read_csv(cancer_dataset_path, encoding='utf-8')

# Preview the initial records of the dataset
print(cancer_data.head())
```

```python
# Preprocess diagnosis labels
cancer_data['diagnosis'] = cancer_data['diagnosis'].map({'M': 1, 'B': 0})

# Extract features and target variable
feature_columns = cancer_data.drop(['id', 'diagnosis', 'Unnamed: 32'], axis=1).values
target_labels = cancer_data['diagnosis'].values

# Normalize feature data
normalizer = StandardScaler()
features_normalized = normalizer.fit_transform(feature_columns)

# Convert to PyTorch tensors
features_tensor = torch.tensor(features_normalized, dtype=torch.float32)
labels_tensor = torch.tensor(target_labels, dtype=torch.float32).view(-1, 1)

# Create train-validation split
features_train, features_val, labels_train, labels_val = train_test_split(
    features_tensor, labels_tensor, test_size=0.2, random_state=42
)

# Prepare data loaders
train_data = TensorDataset(features_train, labels_train)
val_data = TensorDataset(features_val, labels_val)

batch_size = 64
train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_data, batch_size=batch_size)

class BreastCancerClassifier(nn.Module):
    def __init__(self):
        super(BreastCancerClassifier, self).__init__()
        self.hidden_layer = nn.Linear(features_train.shape[1], 32)
        self.activation = nn.ReLU()
        self.output_layer = nn.Linear(32, 1)
        self.probability = nn.Sigmoid()
        
    def forward(self, x):
        hidden = self.activation(self.hidden_layer(x))
        output = self.probability(self.output_layer(hidden))
        return output

def train_classifier(num_epochs):
    """
    Train and validate the breast cancer classifier
    """
    training_start = time.time()
    epoch_train_losses = []
    epoch_val_losses = []
    
    for epoch in range(num_epochs):
        # Training phase
        classifier.train()
        epoch_train_loss = 0
        
        for batch_features, batch_labels in train_loader:
            optimizer.zero_grad()
            predictions = classifier(batch_features)
            batch_loss = loss_function(predictions, batch_labels)
            batch_loss.backward()
            optimizer.step()
            epoch_train_loss += batch_loss.item() * batch_features.size(0)
            
        epoch_train_loss /= len(train_data)
        epoch_train_losses.append(epoch_train_loss)
        
        # Validation phase
        classifier.eval()
        epoch_val_loss = 0
        correct_predictions = 0
        total_samples = 0
        
        with torch.no_grad():
            for batch_features, batch_labels in val_loader:
                predictions = classifier(batch_features)
                batch_loss = loss_function(predictions, batch_labels)
                epoch_val_loss += batch_loss.item() * batch_features.size(0)
                
                predicted_labels = predictions.round()
                total_samples += batch_labels.size(0)
                correct_predictions += (predicted_labels == batch_labels).sum().item()
                
        epoch_val_loss /= len(val_data)
        epoch_val_losses.append(epoch_val_loss)
        validation_accuracy = 100 * correct_predictions / total_samples
        
        print(f'Epoch {epoch+1}/{num_epochs}:')
        print(f'Training Loss: {epoch_train_loss:.4f}')
        print(f'Validation Loss: {epoch_val_loss:.4f}')
        print(f'Validation Accuracy: {validation_accuracy:.2f}%\n')
    
    training_duration = time.time() - training_start
    print(f'Total training time: {training_duration:.2f} seconds')
    
    return epoch_train_losses, epoch_val_losses, training_duration

# Initialize model components
classifier = BreastCancerClassifier()
loss_function = nn.BCELoss()
optimizer = optim.Adam(classifier.parameters(), lr=0.001)

# Execute training
num_epochs = 100
training_losses, validation_losses, training_time = train_classifier(num_epochs)

# Visualize learning curves
plt.figure(figsize=(10, 5))
plt.plot(training_losses, label='Training Loss', color='blue')
plt.plot(validation_losses, label='Validation Loss', color='red')
plt.title('Learning Curves: Training and Validation Loss')
plt.xlabel('Epoch Number')
plt.ylabel('Loss Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Extract final performance metrics
last_epoch_train_loss = training_losses[-1]
last_epoch_val_loss = validation_losses[-1]

# Evaluate final model performance
classifier.eval()
correct_predictions = 0
total_samples = 0

# Calculate final validation metrics
with torch.no_grad():
    for batch_features, batch_labels in val_loader:
        predictions = classifier(batch_features)
        predicted_labels = predictions.round()  # Threshold probabilities at 0.5
        total_samples += batch_labels.size(0)
        correct_predictions += (predicted_labels == batch_labels).sum().item()

final_validation_accuracy = 100 * correct_predictions / total_samples

# Display final performance metrics
print("\nFinal Model Performance Metrics:")
print("-" * 30)
print(f"Training Duration: {training_time:.2f} seconds")
print(f"Final Training Loss: {last_epoch_train_loss:.4f}")
print(f"Final Validation Loss: {last_epoch_val_loss:.4f}")
print(f"Final Validation Accuracy: {final_validation_accuracy:.2f}%")
```

```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import time

# Define the cancer dataset location
cancer_dataset_location = 'cancer (2).csv'

# Load cancer diagnosis data into pandas dataframe
cancer_df = pd.read_csv(cancer_dataset_location, encoding='utf-8')

# Display initial records for data verification
print(cancer_df.head())
```

```python
# Initial data exploration
print("Initial data preview:", cancer_df.head())
print("Column names:", cancer_df.columns.tolist())

# Data cleaning - remove unnecessary column
if 'Unnamed: 32' in cancer_df.columns:
   cancer_df = cancer_df.drop(['Unnamed: 32'], axis=1)

# Encode diagnosis labels
cancer_df['diagnosis'] = cancer_df['diagnosis'].map({'M': 1, 'B': 0})

# Validate data quality
print("Missing values per column:\n", cancer_df.isnull().sum())

# Feature extraction
features = cancer_df.drop(['id', 'diagnosis'], axis=1).values
labels = cancer_df['diagnosis'].values

# Feature normalization 
feature_normalizer = StandardScaler()
features_normalized = feature_normalizer.fit_transform(features)

# Convert to PyTorch tensors
features_tensor = torch.tensor(features_normalized, dtype=torch.float32)
labels_tensor = torch.tensor(labels, dtype=torch.float32).view(-1, 1)

# Create train/validation split
features_train, features_val, labels_train, labels_val = train_test_split(
   features_tensor, labels_tensor, test_size=0.2, random_state=42
)

# Prepare data loaders
train_data = TensorDataset(features_train, labels_train)
val_data = TensorDataset(features_val, labels_val)

batch_size = 64
train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_data, batch_size=batch_size)

class DeepCancerClassifier(nn.Module):
   def __init__(self):
       super(DeepCancerClassifier, self).__init__()
       self.input_layer = nn.Linear(features_train.shape[1], 32)
       self.hidden_layer1 = nn.Linear(32, 64)
       self.hidden_layer2 = nn.Linear(64, 16)
       self.output_layer = nn.Linear(16, 1)
       self.activation = nn.ReLU()
       self.probability = nn.Sigmoid()

   def forward(self, x):
       x = self.activation(self.input_layer(x))
       x = self.activation(self.hidden_layer1(x))
       x = self.activation(self.hidden_layer2(x))
       x = self.probability(self.output_layer(x))
       return x

def train_deep_classifier(num_epochs):
   """
   Train and validate the deep neural network classifier
   """
   training_start = time.time()
   epoch_train_losses = []
   epoch_val_losses = []
   best_accuracy = 0

   for epoch in range(num_epochs):
       # Training phase
       classifier.train()
       epoch_train_loss = 0
       
       for batch_features, batch_labels in train_loader:
           optimizer.zero_grad()
           predictions = classifier(batch_features)
           batch_loss = loss_function(predictions, batch_labels)
           batch_loss.backward()
           optimizer.step()
           epoch_train_loss += batch_loss.item() * batch_features.size(0)
           
       epoch_train_loss /= len(train_data)
       epoch_train_losses.append(epoch_train_loss)

       # Validation phase
       classifier.eval()
       epoch_val_loss = 0
       correct_predictions = 0
       total_samples = 0
       
       with torch.no_grad():
           for batch_features, batch_labels in val_loader:
               predictions = classifier(batch_features)
               batch_loss = loss_function(predictions, batch_labels)
               epoch_val_loss += batch_loss.item() * batch_features.size(0)
               predicted_labels = predictions.round()
               total_samples += batch_labels.size(0)
               correct_predictions += (predicted_labels == batch_labels).sum().item()

       epoch_val_loss /= len(val_data)
       epoch_val_losses.append(epoch_val_loss)
       validation_accuracy = 100 * correct_predictions / total_samples
       best_accuracy = max(best_accuracy, validation_accuracy)

       print(f'Epoch {epoch+1}/{num_epochs}:')
       print(f'Training Loss: {epoch_train_loss:.4f}')
       print(f'Validation Loss: {epoch_val_loss:.4f}')
       print(f'Validation Accuracy: {validation_accuracy:.2f}%\n')

   training_duration = time.time() - training_start
   print(f'Total training time: {training_duration:.2f} seconds')
   
   return epoch_train_losses, epoch_val_losses, training_duration, best_accuracy

# Initialize model components
classifier = DeepCancerClassifier()
loss_function = nn.BCELoss()
optimizer = optim.Adam(classifier.parameters(), lr=0.001)

# Execute training
num_epochs = 100
training_losses, validation_losses, training_time, best_accuracy = train_deep_classifier(num_epochs)

# Visualize learning curves
plt.figure(figsize=(10, 5))
plt.plot(training_losses, label='Training Loss', color='blue')
plt.plot(validation_losses, label='Validation Loss', color='red')
plt.title('Deep Model Learning Curves')
plt.xlabel('Epoch Number')
plt.ylabel('Loss Value')
plt.legend()
plt.grid(True)
plt.show()

# Display final metrics
final_train_loss = training_losses[-1]
final_val_loss = validation_losses[-1]
print("\nFinal Model Performance:")
print("-" * 25)
print(f"Training Duration: {training_time:.2f} seconds")
print(f"Final Training Loss: {final_train_loss:.4f}")
print(f"Final Validation Loss: {final_val_loss:.4f}")
print(f"Best Validation Accuracy: {best_accuracy:.2f}%")
```

```python
# Load CIFAR-10 training dataset
cifar_train_data = datasets.CIFAR10(
    root='./data', 
    train=True, 
    download=True, 
    transform=transforms.ToTensor()
)

# Stack images for statistical calculations
stacked_images = torch.stack([image_tensor for image_tensor, _ in cifar_train_data], dim=3)

# Calculate channel-wise mean and standard deviation
channel_means = stacked_images.view(3, -1).mean(dim=1)
channel_stdevs = stacked_images.view(3, -1).std(dim=1)
```

```python
# Define image transformations pipeline
image_transforms = transforms.Compose([
   transforms.ToTensor(),
   transforms.Normalize(
       mean=(0.4914, 0.4822, 0.4465),
       std=(0.2470, 0.2435, 0.2616)
   )
])
```

```python
# Load test dataset
cifar_test_dataset = datasets.CIFAR10(
   root='./data', 
   train=False, 
   download=False,
   transform=image_transforms
)
```

```python
# Verify image dimensions
sample_image, sample_label = cifar_train_dataset[0]
print(f"Image tensor shape: {sample_image.shape}")
```

```python
# Create data loaders
train_loader = DataLoader(
   cifar_train_dataset,
   batch_size=32,
   shuffle=True,
   num_workers=2
)
```

```python

test_loader = DataLoader(
   cifar_test_dataset,
   batch_size=32,
   shuffle=False,
   num_workers=2
)
```

```python
# Define neural network architecture
classifier = nn.Sequential(
   nn.Flatten(),
   nn.Linear(32 * 32 * 3, 512),  # Flatten RGB image
   nn.Tanh(),                     # Activation function
   nn.Linear(512, 10)            # Output layer for 10 classes
).to(device)

# Initialize training components
loss_function = nn.CrossEntropyLoss()
optimizer = optim.SGD(classifier.parameters(), lr=0.001)
```

```python
def train_classifier(classifier, train_loader, test_loader, num_epochs=300, learning_rate=0.001):
   """
   Train and evaluate a neural network classifier with periodic performance monitoring
   """
   # Setup compute device
   compute_device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
   classifier.to(compute_device)

   # Initialize training components
   loss_function = nn.CrossEntropyLoss()
   optimizer = optim.SGD(classifier.parameters(), lr=learning_rate)

   for epoch in range(num_epochs):
       # Training phase
       classifier.train()
       epoch_loss = 0.0
       
       for batch_inputs, batch_labels in train_loader:
           # Move batch to compute device
           batch_inputs = batch_inputs.to(compute_device)
           batch_labels = batch_labels.to(compute_device)
           
           # Forward and backward passes
           optimizer.zero_grad()
           predictions = classifier(batch_inputs)
           batch_loss = loss_function(predictions, batch_labels)
           batch_loss.backward()
           optimizer.step()
           
           epoch_loss += batch_loss.item()

       # Evaluation phase
       classifier.eval()
       correct_predictions = 0
       total_samples = 0
       prediction_history = []
       label_history = []

       with torch.no_grad():
           for batch_inputs, batch_labels in test_loader:
               # Move batch to compute device
               batch_inputs = batch_inputs.to(compute_device)
               batch_labels = batch_labels.to(compute_device)
               
               # Generate predictions
               predictions = classifier(batch_inputs)
               _, predicted_labels = torch.max(predictions.data, 1)
               
               # Update metrics
               total_samples += batch_labels.size(0)
               correct_predictions += (predicted_labels == batch_labels).sum().item()
               
               # Store predictions and labels for analysis
               prediction_history.extend(predicted_labels.cpu().numpy())
               label_history.extend(batch_labels.cpu().numpy())

       # Calculate accuracy
       epoch_accuracy = correct_predictions / total_samples

       # Display periodic updates
       if epoch % 1 == 0:
           average_loss = epoch_loss / len(train_loader)
           print(f'Epoch {epoch+1}/{num_epochs}:')
           print(f'Average Loss: {average_loss:.4f}')
           print(f'Test Accuracy: {epoch_accuracy * 100:.2f}%\n')

   # Generate final performance report
   performance_report = classification_report(label_history, prediction_history)
   print("\nFinal Classification Report:")
   print("-" * 30)
   print(performance_report)
```

```python
train_classifier(
    classifier=classifier,
    train_loader=train_loader,
    test_loader=test_loader,
    num_epochs=100,
    learning_rate=0.01
)
```

```python
deep_classifier = nn.Sequential(
   nn.Flatten(),                          # Convert image to 1D tensor
   nn.Linear(32 * 32 * 3, 512),          # Input layer
   nn.Tanh(),                            # Activation function
   nn.Linear(512, 256),                  # First hidden layer
   nn.Tanh(),                            # Activation function  
   nn.Linear(256, 128),                  # Second hidden layer
   nn.Tanh(),                            # Activation function
   nn.Linear(128, 10)                    # Output layer for 10 classes
).to(device)
```

```python
train_classifier(
   classifier=deep_classifier,
   train_loader=train_loader,
   test_loader=test_loader,
   num_epochs=100,
   learning_rate=0.01
)
```
